```python
df = pd.read_excel('animali_pulito.ods', index_col=0, engine='odf')  
for col in df.columns:
    if 'Unnamed' in col or ":" in df[col].values:
        del df[col]
df = df.T
del df['INDIC_EF (Labels)']
df = df.dropna(how='any',axis=0)
df['Cows'] = 0

for col in ['head: Cattle - total', 'head: Bovine <1 year old - total', 'head: Bovine 1-<2 years - males', 'head: Bovine 1-<2 years - females', 'head: Bovine 2 years and older - males', 'head: Heifers, 2 years and older', 'head: Dairy cows']:
    # print(df[col].values)
    df['Cows'] += list(map(int, df[col].values))
    del df[col]
df = df.rename(columns={'head: Pigs': 'Pigs',
                        'head: Equidae': 'Horses',
                        'head: Sheep - total': "Sheeps",
                        'head: Goats' : 'Goats',
                        'head: Rabbits (breeding females)': 'Rabbits',
                        '1000 heads: Poultry - broilers': 'K Chickens'})
df["Country"] = ["Bulgaria", "Czechia", "Denmark", "Estonia", "Ireland", "Greece", "Spain", "France", "Croatia", "Italy", "Cyprus",
"Latvia", "Lithuania", "Luxembourg", "Hungary", "Malta", "Netherlands", "Austria", "Poland", "Portugal", "Romania", "Slovenia",
"Slovakia", "Norway", "United Kingdom"]

mains = ["Italy", "Germany", "France", "Poland", "Spain", "Netherlands"]
others_mean = df[~df["Country"].isin(mains)].mean(axis=0).astype(int)
df = df.T
for country in df.columns:
    if country not in mains:
        del df[country]
df["Others"] = others_mean
df = df.T
df["K Chickens"] *= 1000
del df["Horses"]
del df["Goats"]

df.to_csv("../draggableBubble/animals_per_country.csv")
```
