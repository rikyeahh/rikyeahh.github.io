```python
import pandas as pd
metanoTot = pd.read_excel("CLEAN_metano_totale.ods", engine="odf")
metanoTot.columns = ["Country"] + list(range(2012,2022))
countries = metanoTot["Country"]
list(countries)
```

```python
metanoTot = pd.read_excel("CLEAN_metano_totale.ods", engine="odf")
metanoTot.columns = ["Country"] + list(range(2012,2022))
del metanoTot["Country"]

metanoAgri = pd.read_excel("CLEAN_metano_agricoltura.ods", engine="odf")
metanoAgri.columns = ["Country"] + list(range(2012,2022))
del metanoAgri["Country"]

metanoPerc = (metanoAgri / metanoTot) * 100
metanoPerc["Country"] = countries
metanoPerc = metanoPerc.fillna(0)
metanoPerc.to_csv("CLEAN_metano_perc.csv")

metanoPerc
```
