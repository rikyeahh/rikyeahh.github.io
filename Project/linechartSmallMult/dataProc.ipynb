```python
import pandas as pd
import csv
df = pd.read_csv("CLEAN_nitro_agricoltura.csv")
df.T.to_csv("CLEAN_nitro_agricoltura.csv")
```

```python
with open("CLEAN_nitro_agricoltura.csv") as file:
    lines = [line.rstrip() for line in file]

cols = lines[0].split(',')[2:]
print(cols)
del lines[0:1]
print(lines)

f = open('nitro_agri.csv', 'w+', newline='')
writer = csv.writer(f)
writer.writerow(["year","name","n"])
for line in lines:
    fields = line.split(",")
    for i, col in enumerate(cols):
        if col in ['Italy', 'France', "Spain", "Netherlands", "Poland"]:
            output = [int(fields[0]), col, fields[i+2]]
            print(output)
            writer.writerow(output)
f.close()
```
